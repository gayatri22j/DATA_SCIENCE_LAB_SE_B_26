```python
def factorial(n):
    if not isinstance(n, int):
        return "Input must be an integer"
    elif n < 0:
        return "Factorial is not defined for negative numbers."
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

n = int(input("Enter a number: "))
result = factorial(n)

print("Factorial of", n, "is:", result)
```
